```python
#import pandas
import pandas as pd
```

```python
#read excel file
df = pd.read_excel('Cricket data.xlsx')
```

```python
df
```

```python
# rename multiple columns
df = df.rename(columns = {'NO':'Not_Outs','HS':'Highest_Inns_Score','BF':'Balls_Faced','SR':'Strike_Rate','Mat':'Matches'}) 
```

```python
df
```

```python
#Check null values
df.isnull().any()
```

```python
#to fetch the null values
df[df['Balls_Faced'].isna()==1]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
#to identify wether the whole data set has duplicates
df.duplicated()
```

```python
#to identify the duplicates values
df[df['Player'].duplicated()]
```

```python
#To see all the duplicate values
df[df['Player'].isin(('VG Kambli (IND)','DCS Compton (ENG)','V Kohli (IND)','GS Sobers (WI)'))]
```

```python
df[df['Player'].isin(('VG Kambli (IND)','DCS Compton (ENG)','V Kohli (IND)','GS Sobers (WI)'))]
```

```python
#to drop duplicates
df = df.drop_duplicates()
```

```python
df
```

```python
#to split the span value using demiliter '-'
df['Span'].str.split(pat = '-')
```

```python
# to separate the first value from the span column
df['Start_Year']=df['Span'].str.split(pat = '-').str[0]
```

```python
# to separate the second value from the span column
df['End_Year'] = df['Span'].str.split(pat = '-').str[1]
```

```python
df
```

```python
#Drop span column
df = df.drop(['Span'], axis = 1)
```

```python
df.head()
```

```python
# to split country code from player name
df['Country']=df['Player'].str.split(pat = '(').str[1]
```

```python
df['Country']=df['Country'].str.split(')').str[0]
```

```python
df['Country']
```

```python
# fixing player column
df['Player']=df['Player'].str.split(pat = '(').str[0]
```

```python
df.head()
```

```python
#Check data type
df.dtypes
```

```python
# Using asterisk value separating the values into twocolumn
df['Status'] = df['Highest_Inns_Score'].astype(str).str.split(pat ='*').str[1]
```

```python
df['Status'].head()
```

```python
# Now we can use if condition to assign, when it has value consider as 'Not out' else as 'Out' 
def status(Status):
    if pd.isna(Status):
        return 'Out'
    else:
        return 'Not out'
```

```python
df['Status'] = df['Status'].apply(status)
```

```python
df['Status'].head()
```

```python
#To remove asterisk value from Highest_Inns_Score column
df['Highest_Inns_Score']=df['Highest_Inns_Score'].astype(str).str.replace('*', '', regex=False)
```

```python
df['Highest_Inns_Score']
```

```python
df.dtypes
```

```python
#to remove '+' value from columns
df['4s'] = df['4s'].astype(str).str.replace('+','',regex = False)
```

```python
df['Balls_Faced']=df['Balls_Faced'].astype(str).str.replace('+','',regex = False)
```

```python
df['6s']=df['6s'].astype(str).str.replace('+','',regex = False)
```

```python
#To change datatype
df['Highest_Inns_Score']=df['Highest_Inns_Score'].astype(int)
```

```python
df['4s']=df['4s'].astype(int)
```

```python
df['6s']=df['6s'].astype(int)
```

```python
df = df.astype({'Start_Year':'int','End_Year':'int'})
```

```python
#To identify career length
df['Career_Length'] = df['End_Year'] - df['Start_Year']
```

```python
#Find average career length of cricketers
df['Career_Length'].mean()
```

```python
#avg strike rate for cricketers who played over 10 years
df[df['Career_Length']>10]['Strike_Rate'].mean()
```

```python
#find number of cricketers who played before 1960
df[df['Start_Year']<1960]['Player'].count()
```

```python
df.head()
```

```python
#max highest inns by country and sort by desc
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highest_Score').reset_index().sort_values('Highest_Score',ascending = False)
```
