```python
import pandas as pd
```

```python
#Read file
df = pd.read_excel(r"C:\Users\user\Desktop\Python pandas\Customer Call List.xlsx")
df
```

```python
#Drop duplicates
df = df.drop_duplicates()
df
```

```python
#Drop columns
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
#Removing unwanted values in names
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df
```

```python
#Removing unwanted values in names method 2
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
# Replace function using to clean phone number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
#To convert phonenumber column to str using lambda fucntion
df['Phone_Number']= df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
# To seperate phone number with '-' using lamba function
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+ '-'+ x[3:6] + '-'+ x[6:10])
df
```

```python
#Cleaning unwanted values from Phone_NUmber
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
```

```python
df
```

```python
df[["Street_Address","State","Zip_Code"]] = df['Address'].str.split(',',n=2, expand = True)
```

```python
#Cleaning other columns
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('No','N')
```

```python
#Removing null values and n/a value in entire table
df = df.replace('N/a','')
df = df.fillna('')
df
```

```python
#Drop rows where Do_Not_Contact is Y
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
```

```python
#Drop rows which doesn't have phone no
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
```

```python
#Reset index and drop the extra one
df.reset_index(drop = True)
```

```python
#Another method to drop row values which have null
df.dropna(subset = "Phone_Number", inplace = True)
```
